```r
library("tidyverse")
library("patchwork")
set.seed(0)
```

## Diff mean: 20, SD_x: 10, SD_y: 10

```r
x1 <- rnorm(20, mean = 0, sd = 10)
x2 <- rnorm(20, mean = 0, sd = 10)
x3 <- rnorm(20, mean = 0, sd = 10)
x4 <- rnorm(20, mean = 0, sd = 10)
x5 <- rnorm(20, mean = 0, sd = 10)


y1 <- rnorm(20, mean = 20, sd = 10)
y2 <- rnorm(20, mean = 20, sd = 10)
y3 <- rnorm(20, mean = 20, sd = 10)
y4 <- rnorm(20, mean = 20, sd = 10)
y5 <- rnorm(20, mean = 20, sd = 10)


dat <- data.frame(
  x1 = c(x1, y1),
  x2 = c(x2, y2),
  x3 = c(x3, y3),
  x4 = c(x4, y4),
  x5 = c(x5, y5),
  group = c(rep("x", 20), rep("y", 20))
)
```

```r
dat %>% head()

dat %>% tail()
```

```r
ggpubr::ggboxplot(dat, x="group", y="x1", add = "jitter", color = "group") +
  ggpubr::ggboxplot(dat, x="group", y="x2", add = "jitter", color = "group") +
  ggpubr::ggboxplot(dat, x="group", y="x3", add = "jitter", color = "group") +
  ggpubr::ggboxplot(dat, x="group", y="x4", add = "jitter", color = "group") +
  ggpubr::ggboxplot(dat, x="group", y="x5", add = "jitter", color = "group")
```

```r
plot_pca <- function(dat_df) {
    pca_obj <- prcomp(dat_df)
    pca_eigen_val <- round((summary(pca_obj)$importance[2,] * 100),1)
    
    pca_scores <- pca_obj$x %>% data.frame()
    pca_scores$group <- dat$group
    
    pcaplot_pc1_pc2 <- ggplot(pca_scores, aes(x=PC1, y=PC2, color=group)) + 
      geom_point(size=2, position = "jitter") +
      coord_fixed() + 
      ggrepel::geom_text_repel(label=pca_scores$group) +
      xlab(paste0("PC1: ", pca_eigen_val[1], "%")) + 
      ylab(paste0("PC2: ", pca_eigen_val[2], "%")) +
      ggtitle("PCA plot")  +
      theme_classic()
    print(pcaplot_pc1_pc2)
}
```

```r
dat_filt <- dat %>% select(-group)
plot_pca(dat_filt)
```
